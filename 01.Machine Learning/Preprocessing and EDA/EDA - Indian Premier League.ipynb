# Exploratory Analysis and Visualization

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.color_palette("Paired")
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (12, 8)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

### The team with the most number of wins per season.

```python
teams_per_season = data.groupby('season')['winner'].value_counts()
teams_per_season
```

```python
"""
for i, v in win_per_season.iteritems():
    print(i, v)
    
for items in win_per_season.iteritems():
    print(items)    
"""
year = 2008
win_per_season_data = pd.DataFrame(columns=['year', 'team', 'wins'])
for items in teams_per_season.iteritems():    
    if items[0][0]==year:
        print(items)
        win_series = pd.DataFrame({
            'year': [items[0][0]],
            'team': [items[0][1]],
            'wins': [items[1]]
        })
        win_per_season_data = win_per_season_data.append(win_series)
        year += 1
```

```python
win_per_season_data
```

### The number of wins is a discrete value. Hence, we will plot a bar chart(barplot in Seaborn)

```python
sns.barplot('wins', 'team', hue='year', data=win_per_season_data, palette='Paired');
```

### Insight: Mumbai Indians has secured the most wins in four seasons(2010, 2013, 2017, and 2019).

### The venue that hosted the maximum number of matches

```python
venue_ser = data['venue'].value_counts()
```

```python
venue_data = pd.DataFrame(columns=['venue', 'matches'])
for items in venue_ser.iteritems():
    temp_data = pd.DataFrame({
        'venue':[items[0]],
        'matches':[items[1]]
    })
    venue_data = venue_data.append(temp_data, ignore_index=True)
```

```python
plt.title("IPL Venues")
sns.barplot(x='matches', y='venue', data=venue_data);
```

```python
venue_data
```

### Insight:

1.Eden Gardens has hosted the maximum number of IPL matches followed by Wankhede Stadium and M Chinnaswamy Stadium.

2.Till 2019, IPL matches were hosted by 40 venues.

### The most successful IPL team

```python
team_wins_ser = data['winner'].value_counts()

team_wins_data = pd.DataFrame(columns=["team", "wins"])
for items in team_wins_ser.iteritems():
    temp_data = pd.DataFrame({
        'team':[items[0]],
        'wins':[items[1]]
    })
    team_wins_data = team_wins_data.append(temp_data, ignore_index=True)
```

```python
team_wins_data
```

```python
plt.title("Total Victories of IPL Teams")
sns.barplot(x='wins', y='team', data=team_wins_data, palette='Paired');
```

### Insight: Mumbai Indians is the most successful team(as they have won the maximum number of IPL matches -109) followed by Chennai Super Kings and Kolkata Knight Riders.

### Most Valuable Player

```python
mvp_ser = data['player_of_match'].value_counts()

mvp_ten_data = pd.DataFrame(columns=["player", "wins"])
count = 0
for items in mvp_ser.iteritems():
    if count>9:
        break
    else:
        temp_data2 = pd.DataFrame({
            'player':[items[0]],
            'wins':[items[1]]
        })
        mvp_ten_data = mvp_ten_data.append(temp_data2, ignore_index=True)
        count += 1
```

```python
mvp_ten_data
```

```python
plt.title("Top Ten IPL Players")
sns.barplot(x='wins', y='player', data=mvp_ten_data, palette='Paired');
```

### Insight:

1. Cris Gayle is the player who won the most player of the match awards and hence is the most valuable player.

2. Six Indian players have figured in the top ten IPL players list.

### The team that won the most number of toss

```python
toss_ser = data['toss_winner'].value_counts()

toss_data = pd.DataFrame(columns=["team", "wins"])

for items in toss_ser.iteritems():
    temp_data3 = pd.DataFrame({
        'team':[items[0]],
        'wins':[items[1]]
    })
    toss_data = toss_data.append(temp_data3, ignore_index=True)
```

```python
toss_data
```

```python
plt.title("How IPL Teams fared in toss?")
sns.barplot(x='wins', y='team', data=toss_data, palette='Paired');
```

# Insight:

1.Mumbai Indians has won the most toss(till 2019) in IPL history.

2.All the top teams in IPL are successful in winning the toss as well.
