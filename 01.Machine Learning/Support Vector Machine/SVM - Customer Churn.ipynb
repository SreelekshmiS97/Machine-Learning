```python
data['TotalCharges'] = data['TotalCharges'].replace(' ', np.nan).astype(float).values
```

```python
data['TotalCharges']=data['TotalCharges'].fillna(data['TotalCharges'].mean())
```

```python
data['SeniorCitizen']=data['SeniorCitizen'].map({0:'No',1:'Yes'})
```

```python
data.head()
```

```python
data.dtypes
```

```python
X=data.drop(['customerID','Churn'],axis=1)
y=data['Churn']
```

```python
X=pd.get_dummies(X)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=42) #0.67 data will be for training.
```

```python
from sklearn.svm import SVC  #SVC - Supoprt Vector Classifier
cls=SVC()
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'C':[0.1,10], 'gamma':[0.1,0.01],'kernel':['rbf']}
```

```python
grid=GridSearchCV(cls, parameters,cv=5)
grid.fit(X_train, y_train)
```

```python
#to get the best parameter for the model
grid.best_params_
```

```python
svclassifier=SVC(C=10,gamma=0.1,kernel='rbf',probability=True)
svclassifier.fit(X_train,y_train)
```

```python
ypred=svclassifier.predict(X_test)
```

```python
#accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```

```python
#confusison matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,ypred)
```
